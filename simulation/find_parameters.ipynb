```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import curve_fit
```

```python
def theta_dd(I, rcm, m, g, theta, theta_d):
    return -rcm*g*m*np.sin(theta)/I

def single_pendulum_ode(state, t, I, rcm, m, g):
    return np.array([
        state[1],
        theta_dd(I, rcm, m, g, state[0], state[1])
    ])

def simulate(t_data, theta_0, theta_d_0, I, rcm, m, g):
    sol = solve_ivp(
        fun=lambda t, y: single_pendulum_ode(y, t, I, rcm, m, g),
        t_span=(t_data[0], t_data[-1]),
        y0=np.array([theta_0, theta_d_0]),
        t_eval=t_data # Crucial: this forces the solver to output at your exact data timestamps
    )

    theta_sim = sol.y[0]

    return theta_sim
```

```python
t = np.arange(0, 10, 0.01)
hist = simulate(np.arange(0, 10, 0.01), 0.5, 0, 1/3, 0.5, 1, 9.81)
```

```python
res, pcov = curve_fit(simulate, t, hist, p0=[0.3, 0, 1/3, 0.5, 1, 9.81])

res, np.sqrt(np.diag(pcov))
```
